```python
from scripts.tratamento import Tratamento
from scripts.selecao_kbest import SelecaoKBest
from scripts.selecao_chi2 import SelecaoChi2
from scripts.selecao_wrapper import SelecaoWrapper
from scripts.selecao_embedded import SelecaoEmbedded
import pandas as pd
```

```python
org = '../data/matricula.csv'
dst = '../data/out_tecnico_brasil.csv'
#cursos = ['Bacharelado', 'Licenciatura','Tecnologia']
cursos = ['TÃ©cnico']
instituicao = []
qtdCampos = 15
campos = [
    'categoria_situacao',
    'ano',
    'carga_horaria',
    'carga_minima',
#     'cod_matricula',
    'cor_raca',
    'fim_ciclo',
    'inicio_ciclo',
    'data_matricula',
    'eixo_tecnologico',
    'fator_esforco_curso',
    'mes_ocorrencia',
    'modalidade_ensino',
    'nome_curso',
    'fonte_financiamento',
    'renda_familiar',
    'instituicao',
    'sexo',
#     'situacao_matricula',
    'sub_eixo_tecnologico',
    'tipo_curso',
    'tipo_oferta',
    'total_inscritos',
    'turno',
    'codigo_unidade_ensino_sistec',
    'vagas_ofertadas',
    'idade',
    'faixa_etaria',
#    'codigo_ciclo_matricula',
    'unidade_ensino',
    'codigo_municipio_dv',
    'municipio',
    'uf',
    'regiao'
]
t = Tratamento(
    org = org,
    dst = dst,
    tCursos = cursos,
    campos = campos
)
t.instituicao = instituicao
t.processar()
```

```python
seletores = {
    'KBest':SelecaoKBest,
    'Quit2':SelecaoChi2,
    'Wrapper':SelecaoWrapper,
    'Embedded':SelecaoEmbedded
}
selecionados = {}
for nome, seletor in seletores.items():
    sl = seletor(dst,'categoria_situacao')
    sl.balanceada = True
    sl.qtdCampos = qtdCampos
    sl.iniciar()
    selecionados[nome] = pd.Series(sl.selecionados)
df = pd.DataFrame(selecionados)
```
