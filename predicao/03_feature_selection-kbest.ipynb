# Análise features usando KBest
Analisnado as features mais relevantes utilizando o algoritmos KBest

Importando as bibliotecas

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
```

```python
#Desabilitando warning
pd.set_option('mode.chained_assignment',None)
```

### Carregando os dados

```python
matriculas = pd.read_csv("data/out.csv")
```

```python
matriculas.shape
```

```python
y = matriculas[['categoria_situacao']]
```

```python
x = matriculas.iloc[:,range(1, len(matriculas.columns))]
```

### Nomes dos campos

```python
nomes = list(x.columns.values)
```

### Convertendo variáveis categóricas em numéricas

```python
le = LabelEncoder()
for n in nomes: 
    x[n] = le.fit_transform(x[n])
```

### Dimensionando campos com MinMaxScaler

```python
min_max_scaler = MinMaxScaler()
for n in nomes:
    x[[n]] = min_max_scaler.fit_transform(x[[n]])
```

```python
x
```

```python
f_classiftest = SelectKBest(score_func=f_classif,k=15)
```

```python
modelofeat = f_classiftest.fit(x,y.values.ravel())
```

```python
features = modelofeat.transform(x)
```

```python
cols = modelofeat.get_support(indices=True)
list(x.iloc[:,cols].columns)
```
