Análise dos dados oriundos da Plataforma Nilo Peçanha

```r
#Importando dados

matriculas <- read.table("./data/matricula.csv", header=T, sep=",", dec=".", fill = TRUE)
```

### Análise dados

```r
target <- c('categoria_situacao')
categoricos <- c('cor_raca','eixo_tecnologico','modalidade_ensino','nome_curso','fonte_financiamento','renda_familiar','instituicao','sexo','situacao_matricula','sub_eixo_tecnologico','tipo_curso','tipo_oferta','turno','faixa_etaria','unidade_ensino','municipio','uf','regiao')
discretos <- c('ano','carga_horaria','carga_minima','cod_matricula','fim_ciclo','inicio_ciclo','data_matricula','fator_esforco_curso','mes_ocorrencia','total_inscritos','codigo_unidade_ensino_sistec','vagas_ofertadas','idade','codigo_ciclo_matricula')
```

### Sumário dos dados Discretos

```r
for (campo in discretos){
    x <- suppressWarnings(as.numeric(unlist(matriculas[campo])))
    print("-------------------------------------------------------------")
    print(campo)
    print(summary(x))
    print(paste("Desvio Padrão:" , sd(x)))
    
}
```

### Histogramas

```r
hists <-c('ano','carga_horaria','carga_minima','cod_matricula','fator_esforco_curso','total_inscritos','codigo_unidade_ensino_sistec','vagas_ofertadas','idade','codigo_ciclo_matricula')
for (campo in hists){
    x <- suppressWarnings(as.numeric(unlist(matriculas[campo])))
    hist(x,main=campo, xlab=campo)
}
```

### Sumário de dados categóricos

```r
for (campo in categoricos){        
    x <- matriculas[campo]
    barplot(
        table(x),        
        las=3,
        beside = TRUE,             
        main=campo)
}
```
